```python
# Rata % Keseluruhan
mean = df.loc["%per item"].sum()/20
print(round(mean))
```

```python
#Standart Deviasi
stdevi = statistics.stdev(df.iloc[91,6:26].values)
print(round(stdevi))
# print(stdevi)
```

```python
# M-1SD
minsd = round(mean-stdevi)
print(minsd)
```

```python
# M+1SD
plussd = round(mean+stdevi)
print(plussd)
```

```python
# Kesimpulan
def kesimpulan(rata):
    if rata < minsd:
        return "Rendah"
    elif rata < plussd:
        return "Sedang"
    else:
        return "Tinggi"

print(kesimpulan(rata_persen_x1))
print(kesimpulan(rata_persen_x2))
print(kesimpulan(rata_persen_x3))
print(kesimpulan(rata_persen_x4))
print(kesimpulan(rata_persen_y))
```

```python
# Presentase Penerimaan (Kesimpulan Akhir)
kesimpulan_akhir = round((rata_persen_x1 + rata_persen_x2 + rata_persen_x3 + rata_persen_x4 +rata_persen_y)/5)
print("Nilai mean dari Rata % tiap item =",kesimpulan_akhir, "Dikategorikan =", kesimpulan(kesimpulan_akhir))
# print(kesimpulan(kesimpulan_akhir))

def k_akhir(rata):
    if rata < minsd:
        return("Tidak Diterima")
    elif rata < plussd:
        return("Kurang Untuk Diterima")
    elif rata >= plussd:
        return("Diterima")

print(k_akhir(kesimpulan_akhir))
```
